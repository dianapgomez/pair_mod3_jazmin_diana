```python
#¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?
sns.histplot(x = "birth_rate", 
             y= "continent",
             data = df, 
             color = "orange", 
             kde = True, 
             bins = 100 )

plt.xlabel("Tasa de natalidad")
plt.ylabel("Conteo")
plt.title("Tasa de natalidad en el conjunto de datos");
```

```python
#¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? Proporciona una breve descripción de los patrones que observes.

sns.regplot(x = "infant_mortality", 
            y = "life_expectancy", 
            data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1}, # cambiamos el color y el grosor de la linea de tendencia
            scatter_kws = {"color": "blue", "s": 1} # cambiamos el color y el tamaño de los puntos del scaterplot
            )
# cambiamos los nombres de los ejes como hemos estado haciendo hasta ahora
plt.xlabel("Mortalidad infantil")
plt.ylabel("Esperanza de vida")
# ponemos título a la gráfica
plt.title("Relación entre la tasa de mortalidad infantil y la esperanza de vida")
# quitamos la linea de arriba y de la derecha
plt.gca().spines['right'].set_visible(False) # quitamos la línea de la derecha
plt.gca().spines["top"].set_visible(False) # quitamos la línea de arriba;
```
